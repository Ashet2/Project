```python
print('Data Analysis of Postpaid Customers\n')
print('This is a data analysis of postpaid customers of an electricity distribution company.\n',
      'Five different tables are analyzed using Python and its libraries to produce one comprehensive report.\n','The data is analyzed to find the following:\n',
      '1. Explore the data in each table and identify the common columns to merge on\n',
      '2. Extract the relevant columns such as feeder bands and feeder names from 11kV Feeders and 33kV Feeders and then concat the tables.\n',
      '3. Merge the combined feeder list to the postpaid customer list to provided futher information about the customers\n',
      '4. Merge the payment history table to the postpaid customer listing to get the payment pattern of customers.\n',
      '5. Export the final table to a excel file for further analysis.\n')
```

```python
# Importing the libraries and the dataset.

# Importing the libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing the dataset.
postpaid_customers = pd.read_excel('Postpaid Customers Listing.xlsx')
all_public_dts = pd.read_excel('All Public DTs.xlsx')
payment_history = pd.read_excel('Payment History.xlsx')
kv11feeders = pd.read_excel('11kV Feeders.xlsx')
kv33feeders = pd.read_excel('33kV Feeders.xlsx')
```

```python
# Data Exploration: Customer Listing.
print('Postpaid Customers Listing\n', postpaid_customers.info(),'\n')

# Print the first five rows of the dataset.
print('First five rows of the dataset:\n', postpaid_customers.head(),'\n')
```

```python
# Data Exploration: Payment History.
print('Payment History\n', payment_history.info(),'\n')

# Print the first five rows of the dataset.
print('First five rows of the dataset:\n', payment_history.head(),'\n')
```

```python
# Data Exploration: All Public DTs.
print('Payment History\n', all_public_dts.info(),'\n')

# Print the first five rows of the dataset.
print('First five rows of the dataset:\n', all_public_dts.head(),'\n')
```

```python
# Drop duplicate value from the feeder number column.
kv11feeders = kv11feeders.drop_duplicates(subset=['Feeder Number'])
print('First five rows of the dataset:\n', kv11feeders.info(),'\n')
```

```python
# Extract the required columns(Feeder Number, Feeder Names, Band) from the dataset.

required11kvfeeders_column = kv11feeders[['Feeder Number','Feeder Names','SRT BAND']]

# Change the data type of the Feeder Number column to integer.
required11kvfeeders_column['Feeder Number'] = required11kvfeeders_column['Feeder Number'].fillna(0).astype('int64')

print('Required columns from the dataset:\n', required11kvfeeders_column.head(),'\n')
```

```python
# Drop duplicate value from the feeder number column.
kv11feeders = kv11feeders.drop_duplicates(subset=['Feeder Number'])
print('First five rows of the dataset:\n', kv11feeders.info(),'\n')
```

```python
# Data Exploration: 33kV Feeders
print('33kV Feeders\n', kv33feeders.info(),'\n')

# Print the first five rows of the dataset.
print('First five rows of the dataset:\n', kv33feeders.head(),'\n')
```

```python
# Drop duplicate value from the feeder number column.
kv33feeders = kv33feeders.drop_duplicates(subset=['Feeder Number'])
print('First five rows of the dataset:\n', kv33feeders.info(),'\n')
```

```python
# Extract the required columns(Feeder Number, Feeder Names, Band) from the dataset.

required33kvfeeders_column = kv33feeders[['Feeder Number','Feeder Names','SRT BAND']]

# Change the datatypr of feeder number column name to integer.
required33kvfeeders_column['Feeder Number'] = required33kvfeeders_column['Feeder Number'].fillna(0).astype(np.int64)

print('Required columns from the dataset:\n', required33kvfeeders_column.head(),'\n')
```

```python
# Combine rows of the 11kV and 33kV feeders dataset.
feeder_list = pd.concat([required11kvfeeders_column, required33kvfeeders_column])


print('Combined rows of the 11kV and 33kV feeders dataset:\n', feeder_list.head(),'\n')
```

```python
# Merging the postpaid customer listing and the feeder list dataset.
postpaid_customers_feeder_list = postpaid_customers.merge(feeder_list, on= 'Feeder Number', how='left', indicator=True)

# Print the first five rows of the dataset.

print('Postpaid Customers Listing\n', postpaid_customers_feeder_list.info(),'\n')
print('First five rows of the dataset:\n', postpaid_customers_feeder_list.head(),'\n')
print('Comment:\n', 'We have missing data in the SRT BAND column.\n', 'We need to analyse further to identify the feeders with the missing data.','\n')
```

```python
# Filter Feeder Names with NA in the SRT BAND column.
sbt_band_na = postpaid_customers_feeder_list[postpaid_customers_feeder_list['SRT BAND'].isna()]
sbt_band_na = sbt_band_na['Feeder Name'].unique()
print('Feeder Names with NA in the SRT BAND column:\n', sbt_band_na,'\n')
print('Comment:\n', 'We have five feeders with missing data in the SRT BAND column.\n', 
      'We need to analyse further to know why the feeders have missing data.','\n')
```

```python
# Filter out the feeders on the customer listing that are not on the feeder list.
feeder_not_on_feeder_list = postpaid_customers_feeder_list.loc[postpaid_customers_feeder_list['_merge']== 'left_only', 'Feeder Name'].unique()

print('Feeders with missing data:\n', feeder_not_on_feeder_list,'\n')
print('Comment:\n', 'Two feeders are not on the feeder list\n', 'we need to check the feeder list to see if the feeders are on the list.\n',
      'Out of the five feeders with missing data, three are on the feeder list.\n', 
      'We need to check the feeder list to see if the other two feeders are on the list.','\n')
```

```python
# Check if names like Ajegunle Interface in the feeder names column of the feeder list and postpaid customers listing.
fd_aje = feeder_list[feeder_list['Feeder Names'].str.contains(('AJEGUNLE INTERFACE'), na=False)]

fd_aje_cus = postpaid_customers[postpaid_customers['Feeder Name'].str.contains(('AJEGUNLE INTERFACE'), na=False)]
fd_aje_cus = fd_aje_cus[['Feeder Number', 'Feeder Name']]
fd_aje_cus = fd_aje_cus.drop_duplicates()

# Check if names like Fakale Maya in the feeder names column of the feeder list and postpaid customers listing.
fd_maya = feeder_list[feeder_list['Feeder Names'].str.contains(('MAYA'), na=False)]

fd_maya_cus = postpaid_customers[postpaid_customers['Feeder Name'].str.contains(('MAYA'), na=False)]
fd_maya_cus = fd_maya_cus[['Feeder Number', 'Feeder Name']]
fd_maya_cus = fd_maya_cus.drop_duplicates()

# Print the result.
print('Nomenclature and feeder number of Ajegunle on feeder list:\n',fd_aje, '\n')
print('Nomenclature and feeder number of Ajegunle on postpaid customer listing:\n', fd_aje_cus, '\n')
print('Nomenclature and feeder number of Fakale Maya on feeder list:\n', fd_maya, '\n')
print('Nomenclature and feeder number of Fakale Maya on postpaid customer listing\n', fd_maya_cus, '\n')

print('Comment:\n', 'The two missing feeders have mismatched feeder numbers in the two tables.\n',
      'Also, three of the  feeders (33-MarylandTCN-AJEGUNLE INTERFACE, 11-IgbogboINJ-T2-Agunfoye, 11-OdogunyanINJ-T2-Odokekere) does not have srt band in the feeder list.\n')
```

```python
# Update the records of the feeders with mismatched feeder numbers and missing srt bands in the feeder list.

# Update the feeder number and srt band of 33-MarylandTCN-AJEGUNLE INTERFACE
required33kvfeeders_column.loc[71, ['Feeder Number', 'SRT BAND']] = [1103849, 'DIAMOND']

# Update the feeder number of 33-OdogunyanTCN-FAKALE MAYA
required33kvfeeders_column.loc[58, ['Feeder Number']] = [1069475]

# Update the srt band of 11-OdogunyanINJ-T2-Odokekere
required11kvfeeders_column.loc[214, ['SRT BAND']] = ['SILVER']

# Update the srt band of 11-IgbogboINJ-T2-Agunfoye
required11kvfeeders_column.loc[111, ['SRT BAND']] = ['SILVER']
```

```python
# Combine rows of the 11kV and 33kV of the updated required feeders dataset.
feeder_list_updated = pd.concat([required11kvfeeders_column, required33kvfeeders_column], ignore_index=True)

# Drop the Feeder Names column from the feeder list dataset.
feeder_list_updated = feeder_list_updated.drop(['Feeder Names'], axis=1)

# Merging the postpaid customer listing and the updated feeder list dataset.
postpaid_customers_srt_band = postpaid_customers.merge(feeder_list_updated, on= 'Feeder Number', how='left', indicator=True)

# Print the first five rows of the dataset.

print('Postpaid Customers Listing\n', postpaid_customers_srt_band.info(),'\n')
```

```python
# Explore the payment history dataset.
print('Payment History\n', payment_history.info(),'\n')
```

```python
# Drop province and locaion columns from the payment history dataset.
payment_history_requied_columns = payment_history.drop(['Province', 'Location'], axis=1)

# Merge the postpaid customer listing with srt band and the payment history dataset.
postpaid_customers_srt_band_payment_history = postpaid_customers_srt_band.merge(payment_history_requied_columns, on= 'Account No.', how='left')

# Print the first five rows of the dataset.
print('First five rows of the dataset:\n', postpaid_customers_srt_band_payment_history.head(),'\n')

# Print the information of the dataset.
print('Postpaid Customers Listing\n', postpaid_customers_srt_band_payment_history.info(),'\n')

```

```python
# Print summary.
print('Summary:\n', 'We have 129,662 accounts in the postpaid customer listing.\n',
      'All the required data has been merged into one dataset and I have updated all the missing/mismatched data.\n',
      'The merged dataset is ready for further analysis in python or other visulization/analysis tools.\n')
```

```python
# Export the dataset to excel.
postpaid_customers_srt_band_payment_history.to_excel('Postpaid Customers SRT BAND Payment History.xlsx', index=False)
```
